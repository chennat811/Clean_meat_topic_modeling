```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import numpy as np
```

```python
topics_df=pd.read_csv('topic_df_lda_nov12.csv')
```

```python
topics_df['datetime']=pd.to_datetime(topics_df['date'], format = '%Y-%m-%d %H:%M:%S')
```

```python
topics_df['month_year'] = topics_df['datetime'].dt.to_period('M')
```

```python
count = pd.DataFrame(columns = ['month_year','Dominant Topic'], data = topics_df)
df_time_counts = count.groupby(['month_year','Dominant Topic']).size().reset_index(name = 'counts')
```

```python
df_time_counts = pd.DataFrame(df_time_counts)
```

```python
counts = df_time_counts['counts']
```

```python
fig,ax= plt.subplots(figsize=(20,10))

for n, group in df_time_counts.groupby('Dominant Topic'):
    group.plot(x='month_year',y='counts', ax=ax,label=n)
```

```python
df_time_counts.sort_values('counts', ascending = False)
```
